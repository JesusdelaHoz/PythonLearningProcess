**1) Trabajemos con NumPy**

* Importa numpy y usa un alias para poder utilizar sus recursos
* Crea un array de elementos (del 10 al 100)
* Invertir el array
* Crear un array 3x3 con valores de 0 a 8
* Crear un array identidad de 6x6
* Crear un array con valores al azar con forma 3x3x3
* Crear dos arrays con valores al azar A y B, verificar si son iguales (hay varias opciones, investigar)

```python
import numpy as nm

arr100 = list(range(101))
del arr100[:10]

# El array del 10 al 100
miArray = nm.array(arr100)
print(miArray)

# Invertimos el array
arrayInvertido = miArray[::-1]
print(arrayInvertido)

# Creamos el array 3x3
m = nm.arange(0, 9)
final = nm.reshape(m, [3,3])
print(final)

# La matriz identidad de 6x6
identi = nm.identity(6)
print(identi)

# Matriz aleaotria
mRandom = nm.random.randint(100, size = (3, 3, 3))
print(mRandom)

a = nm.random.randint(100, size = (3, 3, 3))
b = nm.random.randint(100, size = (3, 3, 3))
print("Los arrays 'a' y 'b' son iguales --->", nm.array_equiv(a, b))
```

**2) Sigamos con NumPy**

* Crear un array de 100 elementos aleatorios comprendidos entre 1 y 1000 (pueden ser números enteros).
* Este array se llamara dólares
* Crear otro array llamado euros
* Realizar la conversión de los precios en dólares del primer array al segundo array en euros. Se puede serguir la siguiente conversión:
    - 1 EUR = 1,14362 USD
    - 1 USD = 0,874413 EUR
* Mostrar el resultado del array de euros pero redondeado a 2 decimales

```python
dolares = nm.random.randint(1000, size = 100)
euros = nm.zeros(100)
print(euros)

conversionDolares = dolares * 0.874413
euros = conversionDolares + euros
print("\n", euros)

# Redondeo a dos decimales
eurosRound = nm.around(euros, decimals = 2)
print("\n", eurosRound)
```

* Mostrar el número de registros del dataframe de titulos

* Mostrar las 5 peliculas más antiguas del listado de titulos
* Mostrar las peliculas que en el titulo tienen la palabra "Dracula". También mostrar el número total de peliculas que coincidan con este requisito
* Mostrar los 10 titulos más comunes (que más se repiten)
* Mostrar cual fue la primer pelicula hecha titulada "Romeo and Juliet" 
* Listar todas las peliculas que contengan la palabra "Exorcist" ordenadas de la más vieja a la más reciente
* Mostrar cuantas peliculas fueron hechas en el año 1950

* Mostrar todos los roles o papeles que hubo en la pelicula "The Godfather". También mostrar el número total de coincidencias

* Listado de papeles como protagonista (n=1) que tuvo el actor "Charlton Heston" en la década de los 60's, ordenado por año de forma descendente
* Mostrar cuantos papeles para actores hubo en la década de los 50's
* Mostrar cuantos papeles para actrices hubo en la década de los 50's

```python
import pandas as pd
```

```python
# Caragamos los CSV y vemos sus 5 primeros registros
titulos = pd.read_csv("05_05_imdb_titulos.csv")
elenco = pd.read_csv("05_05_imdb_elenco.csv")
print(titulos.head())
print("\n", elenco.head())

# Numero de registros
print("\nEl numero de registros de TITULOS es de", len(titulos), 'registros')
print("\nEl numero de registros de ELENCO es de", len(elenco), 'registros')

# Las peliculas mas antiguas
titulosOrdenYear = titulos.sort_values('year', ascending = True)
print("Las cinco peliculas mas antiguas son las siguientes:\n", titulosOrdenYear.head())

# Las que tengan las palabra Dracula
condicion = titulos["title"].str.find("Dracula")
titulosDracula = titulos["title"][condicion != -1]
print("Las peliculas que contienen la palabra 'Dracula' son las siguientes:")
for i in titulosDracula:
    print("\t-->", i)
print("Un total de {} titulos".format(len(titulosDracula)))

# Mostrar 10 titulos mas repetitivos
rep10titulos = titulos["title"].value_counts().head(10)
print(rep10titulos)

#La pelicula primigenia de "Romeo y Julieta"
romJul = titulosOrdenYear.loc[titulosOrdenYear["title"] == "Romeo and Juliet", ["title", "year"]]
print(romJul.head(1))

# Ordenar las que contengan "Exorcist"
condicion = titulos["title"].str.find("Exorcist")
dtexorTiYear = titulos.loc[condicion != -1, ["title", "year"]]
exorOrdYear = dtexorTiYear.sort_values('year', ascending = True)
print(exorOrdYear)

# La cantidad de peliculas hechas en 1950
titulos1950 = titulos["title"][titulos["year"] == 1950]
print("El numero de peliculas hechas en '1950' es de {} peliculas".format(len(titulos1950)))

# Hechas entre 1950 y 1959
titulos5059 = titulos["title"][(titulos["year"] >= 1950) & (titulos["year"] <= 1959)]
print("El numero de peliculas hechas entre los años 50 y 59 es de {} peliculas".format(len(titulos5059)))


# Roles en "The Godfather"
rolesPadrin = elenco.loc[(elenco["title"] == "The Godfather"), ["character"]]
print("Los roles son los siguientes:")
for i in rolesPadrin["character"]:
    print("\t--->", i)
    
rolesDupli = rolesPadrin.value_counts()
contador = 0
for r in rolesDupli:
    if r > 1:
        contador += 1
print(contador, "coincidencias en los roles")


# Elenco ordenado por n 
dtDracula1958 = elenco.loc[(elenco["title"] == "Dracula") & (elenco["year"] == 1958), ["year", "title", "character", "n"]]
dtDracula1958.sort_values('n')

# Bruce Wayne a lo largo de la historia
numBW_historia = len(elenco["character"][elenco["character"] == "Bruce Wayne"])
print("Las apariciones en peliculas de 'Bruce Wayne' han sido", numBW_historia)

# Numeros de papeles en la carrera de Robert
numRob = len(elenco["name"][elenco["name"] == "Robert De Niro"])
print("El numero de papeles en la carrera de Robert de Niro es de", numRob)

# Papeles de protagonista de "Charlton Heston" en la decada de los 60´s
dfPapeleProtaChar = elenco.loc[(elenco["year"] >= 1960) & 
           (elenco["year"] <= 1969) & 
           (elenco["name"] == "Charlton Heston") & 
           (elenco["n"] == 1), 
           ["name", "year", "character", "n"]]
print(dfPapeleProtaChar.sort_values('year', ascending = False))

# Papeles actores 50´s
serPapelM50 = elenco["type"][(elenco["type"] == "actor") & 
                            (elenco["year"] >= 1950) & 
                            (elenco["year"] <= 1959)]
print("El numero de papeles para actores en las decada de los 50´s es de", len(serPapelM50))

# Papeles actrices 50´s
serPapelF50 = elenco["type"][(elenco["type"] == "actress") & 
                            (elenco["year"] >= 1950) & 
                            (elenco["year"] <= 1959)]
print("El numero de papeles para actrices en las decada de los 50´s es de", len(serPapelF50))
```
