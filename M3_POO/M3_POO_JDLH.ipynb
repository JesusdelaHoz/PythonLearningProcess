#### Creación

Sobre el ejercicio anterior realizar la siguiente modificación.

Si en el ejercicio hacemos un print(alumno1.nota) esto nos devolverá la nota del alumno. Esto puede llegar a ser un problema de seguridad, vamos a encapsular el código para que no podamos acceder directamente a las variables, y el acceso tenga que ser a través de métodos.

* Hacer que las variables de la clase sean __privadas__. Recordatorio: Añadir doble guion delante del nombre de la variable
* Crea métodos __getter y setter__ para poder acceder y modificar las variables de la clase

#### Experimentación

* Crea algunos alumnos
* Comprueba que solo podamos acceder a las variables de la clase a través de sus métodos get y set

````python
class Alumno():
    
    # Constructor de la clase
    def __init__(self, nombre, nota):
        self.__nombre = nombre
        self.__nota = nota
        print("El alumno ha sido creado")
     
    # Ahora creamos los getter y setter de cada variable
    @property
    def nombre(self):
        print("Getter nombre en accion")
        return self.__nombre
    
    @nombre.setter
    def nombre(self, nuevoNombre):
        print("Setter nombre en accion")
        self.__nombre = nuevoNombre
        
    @property
    def nota(self):
        print("Getter nota en accion")
        return self.__nota
    
    @nota.setter
    def nota(self, nuevaNota):
        print("Setter nota en accion")
        self.__nota = nuevaNota
    
    
    # El metodo tambien lo metoemos pero modificado junyo las variables
    
    def calificacion(self):
        if self.__nota < 5:
            return  "El alumno {} esta SUSPENSO".format(self.__nombre)
        else:
            return "El alumno {} esta APROBADO".format(self.__nombre)
        
        
    # Mostrar la informacion
    
    def __str__(self):
        return "El alumno {} ha sido calificado con un {} de nota en el curso".format(self.__nombre, 
                                                                                      self.__nota)

    
    
if __name__ == "__main__":
    
    alumnoEmpollon = Alumno("Jose Luis Romero", 10)
    alumnoMalote = Alumno("Bryan Gonzalez", 3)
    alumnoNormalillo = Alumno("Jesus de la Hoz", 5)

    print(alumnoEmpollon)
    print(alumnoMalote)
    print(alumnoNormalillo)

    alumnoEmpollon.nota = 5
    print(alumnoEmpollon.nota)
    resultado2 = alumnoMalote.calificacion()
    print(resultado2)
    
    print(alumnoEmpollon.__nota) # Comprobamos si es privada
    
````
