**1) Ejercicio de herencia múltiple y polimorfismo:**
* Crea la clase Vehiculos con un constructor que incluya marca y modelo
* Esta clase Vehiculos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar gasolina"
* Crea la clase VElectricos con un constructor que incluya marca, modelo y autonomia
* Esta clase VElectricos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar electricidad"
* Crea la clase BicicletaElectrica que herede de Vehiculos y de VElectricos, pero dando prioridad a VElectricos (ya que es un vehículo más electrico que normal)
* Crea la clase Quad que herede de Vehiculos y de VElectricos, pero dando prioridad a Vehiculos (ya que sólo usa la electricidad de modo puntual)
* Crear un objeto de BicicletaElectrica y otro de Quad
* Emplear las técnicas de polimorfismo aprendidas para conseguir que al peguntarle a los objetos que acabamos de crear por su repostaje, ambos nos respondan adecuadamente



```python
class Vehiculo():
    
    #Constructor
    def __init__(self, marca = None, modelo = None):
        self.__marca = marca
        self.__modelo = modelo
        print("Vehiculo listo")
    
    @property
    def marca(self):
        return self.__marca
    @marca.setter
    def marca(self, newMarca):
        self.__marca = newMarca
        
    @property
    def modelo(self):
        return self.__modelo
    @modelo.setter
    def modelo(self, newModelo):
        self.__modelo = newModelo
    
    # Vamos con los metodos de la clase
    def repostar(self):
        print("Este vehiculo tiene que repostar gasolina")
    
    
    def __str__(self):
        return "Este vehiculo es de la marca {} y modelo {}".format(self.__marca, self.__modelo)

    
class VElectricos():
    
    #Constructor
    def __init__(self, marca = None, modelo = None, autonomia = None):
        self.__marca = marca
        self.__modelo = modelo
        self.__autonomia = autonomia
        print("Vehiculo electrico listo")
    
    @property
    def marca(self):
        return self.__marca
    @marca.setter
    def marca(self, newMarca):
        self.__marca = newMarca
        
    @property
    def modelo(self):
        return self.__modelo
    @modelo.setter
    def modelo(self, newModelo):
        self.__modelo = newModelo
        
    @property
    def autonomia(self):
        return self.__autonomia
    @autonomia.setter
    def autonomia(self, newAutonomia):
        self.__autonomia = newAutonomia
    
    # Vamos con los metodos de la clase
    def repostar(self):
        print("Este vehiculo tiene que repostar electricidad")
    
    
    def __str__(self):
        return "Este vehiculo electrico es de la marca {} , modelo {} y tiene una autonomia de {} km".format(self.__marca, 
                                                                                                   self.__modelo, 
                                                                                                   self.__autonomia)
    
    
    
class BicicletaElectrica(VElectricos, Vehiculo):
    
    # Constructor, ya que proviene de dos clases, eligo que escoger de cada una
    def __init__(self, marca = None, modelo = None , autonomia = None):
        Vehiculo.__init__(self, marca, modelo)
        VElectricos.__init__(self, autonomia = autonomia) 
        
    
    def repostar(self):   #La prioridad de este objeto es la electricidad
        print("Este vehiculo tiene que repostar electricidad")
    
    
    def __str__(self):
        return VElectricos.__str__(self)

    

class Quad(Vehiculo, VElectricos):
    
    # Constructor
    def __init__(self, marca = None, modelo = None , autonomia = None):
        Vehiculo.__init__(self, marca, modelo)
        VElectricos.__init__(self, autonomia = autonomia)
    
    def repostar(self):
        print("Este vehiculo tiene que repostar gasolina y si no enciende, revisar bateria")
    
    # Al ser mas de gasolina que electrico, la autonomia es indiferente, si se quiere saber se preguntara al objeto
    def __str__(self): 
        return Vehiculo.__str__(self)


def repostar(vehiculo):
    vehiculo.repostar()

    
if __name__ == "__main__":
    
    q1 = Quad("Yamaha", "XR250", "10000 h" ) # Al ser bateria solo de encendido, la autonomia la mido en horas
    b1 = BicicletaElectrica("SantaCruz", "29'", "60")
    repostar(q1)
    repostar(b1)
```

```python
class Personal_Universitario():
    
    # Constructor
    def __init__(self, datos = {"id": None , 
                                "nombre": None , 
                                "email": None}):
        self.__datos = datos
        print("Personal universitario registrado")
    
    @property
    def datos(self):
        print("GETTER personal")
        return self.__datos
    @datos.setter               # El setter de un diccionario, se tendria que meter de nuevos todo con la modificacion
    def datos(self, newDatos):
        self.__datos = newDatos
    
    
    
    def mostrarDatos(self):
        print(self.__datos)

    
class Oficina(Personal_Universitario):
    
    def __init__(self, datos, puesto):
        super().__init__(datos)
        self.__puesto = puesto
        self.datos["Puesto"] = self.__puesto
        print("Puesto de oficina creado")
    
    @property
    def puesto(self):
        print("GETTER Oficina")
        return self.__puesto
    @puesto.setter
    def puesto(self, newPuesto):
        self.__puesto = newPuesto
        
    def mostrarDatos(self):
        print(self.datos)
        

class Profesor(Personal_Universitario):
    
    def __init__(self, datos, especial):
        super().__init__(datos)
        self.__especial = especial
        self.datos["Especializacion"] = self.__especial
        print("Puesto de profesor creado")
    
    @property
    def especial(self):
        print("GETTER Profesor")
        return self.__especial
    @especial.setter
    def especial(self, newEspecail):
        self.__especial = newEspecial
        
     
    def mostrarDatos(self):
        print(self.datos)
    
class Alumno(Personal_Universitario):
    
    def __init__(self, datos, creditos = int()):
        super().__init__(datos)
        self.__creditos = creditos
        self.datos["Creditos Aprobados"] = self.__creditos
        print("Alumno ready")
    
    @property
    def creditos(self):
        print("GETTER Alumno")
        return self.__creditos
    @creditos.setter
    def creditos(self, newCreditos):
        self.__creditos = newCreditos
    
    def mostrarDatos(self):
        print(self.datos)

        
        
def mostrarDatos(o):
    o.mostrarDatos()
        
        
if __name__ == "__main__":
    
    ofc = Oficina({"id": 152615 , "nombre": "Roberto Garcia" , "email": "Roberto@.gmail.com"}, "Secretario")
    alum = Alumno({"id": 356298 , "nombre": "Jesus de la Hoz" , "email": "jesusito@.gmail.com"}, 70)
    prf = Profesor({"id": 856987 , "nombre": "Carlos Villa" , "email": "carlitros.villa@.gmail.com"}, "Entomologia")
    persUniv = Personal_Universitario({"id": 457898, "nombre": None,"email": "@.gmail.com"})
    mostrarDatos(ofc)
    mostrarDatos(alum)
    mostrarDatos(prf)
    mostrarDatos(persUniv)
```

```python
# La tematica de las clases sera del tipo de bici, y dependiendo de sus atributos cambiara
# La funcion será una revision antes de utilizar el tipo de bici

class Bicicleta():
    
    # Constructor
    def __init__(self, marca = None, modelo = None, ruedas = None, material = None):
        self.__marca = marca
        self.__modelo = modelo
        self.__ruedas = ruedas
        self.__material = material
        print("Bicicleta lista")
    
    @property
    def marca(self):
        return self.__marca
    @marca.setter
    def marca(self, newMarca):
        self.__marca = newMarca
    
    @property
    def modelo(self):
        return self.__modelo
    @modelo.setter
    def modelo(self, newModelo):
        self.__modelo = newModelo
        
    @property
    def ruedas(self):
        return self.__ruedas
    @ruedas.setter
    def ruedas(self, newRuedas):
        self.__ruedas = newRuedas
        
    @property
    def material(self):
        return self.__material
    @material.setter
    def material(self, newMaterial):
        self.__material = newMaterial
    
    
    def revisionPREuso(self):
        print("Revisar aire de las ruedas")
    
    
    def __str__(self):
        return "La bicicleta es de la marca {}, modelo {} y esta hecha con el material {}".format(self.__marca, 
                                                                                                  self.__modelo, 
                                                                                                  self.__material)

    
    
    
class Bici_Montaña(Bicicleta):
    
    # Constructor
    def __init__(self, 
                 marca = None, 
                 modelo = None, 
                 ruedas = None, 
                 material = None, 
                 cubierta = None, 
                 anchura_rueda = None,
                 amortiguacion_delantera = None, 
                 amortiguacion_trasera = None):
        Bicicleta.__init__(self, marca, modelo, ruedas, material)
        self.__cubierta = cubierta
        self.__anchura_rueda = anchura_rueda
        self.__amortiguacion_delantera = amortiguacion_delantera
        self.__amortiguacion_trasera = amortiguacion_trasera
        print("Bicicleta de montaña lista")
    
    @property
    def cubierta(self):
        return self.__cubierta
    @cubierta.setter
    def cubierta(self, newCubierta):
        self.__cubierta = newCubierta
        
    @property
    def anchura_rueda(self):
        return self.__anchura_rueda
    @anchura_rueda.setter
    def anchura_rueda(self, newAnchura):
        self.__anchura_rueda = newAchura
        
    @property
    def amortiguacion_delantera(self):
        return self.__amortiguacion_delantera
    @amortiguacion_delantera.setter
    def amortiguacion_delantera(self, newAmortiguacionDel):
        self.__amortiguacion_delantera = newAmortiguacionDel
    
    @property
    def amortiguacion_trasera(self):
        return self.__amortiguacion_trasera
    @amortiguacion_trasera.setter
    def amortiguacion_trasera(self, newAmortiguacionTra):
        self.__amortiguacion_trasera = newAmortiguacionTra
    
    
    def revisionPREuso(self):
        print("Revisar aire de las ruedas, tacos de las ruedas y amortiguacion")
    
    
    def __str__(self):
        return Bicicleta.__str__(self) + "\nTiene amortiguacion delantera : {} y trasera: {}".format(self.__amortiguacion_delantera, 
                                                                                                    self.__amortiguacion_trasera)

    
    
class Bici_Carretera(Bicicleta):
    
    # Constructor
    def __init__(self, 
                 marca = None, 
                 modelo = None, 
                 ruedas = None, 
                 material = None, 
                 cubierta = None, 
                 anchura_rueda = None):
        Bicicleta.__init__(self, marca, modelo, ruedas, material)
        self.__cubierta = cubierta
        self.__anchura_rueda = anchura_rueda
        print("Bicicleta de carretera lista")
        
    @property
    def cubierta(self):
        return self.__cubierta
    @cubierta.setter
    def cubierta(self, newCubierta):
        self.__cubierta = newCubierta
    
    @property
    def anchura_rueda(self):
        return self.__anchura_rueda
    @anchura_rueda.setter
    def anchura_rueda(self, newAnchura):
        self.__anchura_rueda = newAchura
    
    def revisionPREuso(self):
        print("Revisar aire de las ruedas y el desgaste del caucho de la rueda")
    
    
    def __str__(self):
        return Bicicleta.__str__(self) + ".Es rigida completamente"

    
    
class Monociclo(Bicicleta):
    
    def __init__(self, marca = None, modelo = None, ruedas = None, material = None):
        super().__init__(marca, modelo, ruedas, material)
        print("Monociclo listo")
    
    def revisionPREuso(self):
        print("Revisar aire de la rueda y sillin")
    
    
class Bici_Descenso(Bici_Montaña):
    
    def revisionPREuso(self):
        print("""Revisar encarecidamente la amortiguacion, delantera y trasera.
        Revisar aire de las ruedas y tacos de las ruedas""")
    

class Bici_Gravel(Bici_Carretera, Bici_Montaña):
    
    # Constructor
    def __init__(self, 
                 marca = None, 
                 modelo = None, 
                 ruedas = None, 
                 material = None, 
                 cubierta = None, 
                 anchura_rueda = None,
                 amortiguacion_delantera = None):
        Bici_Carretera.__init__(self, marca, modelo, ruedas, material, cubierta, anchura_rueda)
        Bici_Montaña.__init__(self, amortiguacion_delantera = amortiguacion_delantera)
        print("GRAVEL lista")
        
        
    def revisionPREuso(self):
        print("Revisar la amortiguacion, delantera. Revisar aire de las ruedas y grid de las ruedas")
        
    
    def __str__(self):
        return "Esta bicicleta tiene una anchura de rueda {}, cubierta {} y amortiguacion delantera {}".format(self.anchura_rueda, 
                                                                                                              self.cubierta, 
                                                                                                              self.amortiguacion_delantera)
        
    
        
    
def revisionPREuso(b):
    b.revisionPREuso()

    
    
if __name__ == "__main__":
    
    mtb = Bici_Montaña(marca = "Cubo", modelo = "Extreme", ruedas = 2, material = "Carbono", 
                       cubierta = "Tacos", anchura_rueda = "Grande", amortiguacion_delantera = True, amortiguacion_trasera = True)
    bCarretera = Bici_Carretera(marca = "Specialez", modelo = "Gool", ruedas = 2, material = "Aleacion de carbono", 
                       cubierta = "Lisa", anchura_rueda = "Fina")
    mono = Monociclo(marca = "Circus", modelo = "Normal", ruedas = 1, material = "Hierro")
    gravel = Bici_Gravel(marca = "Cubo", modelo = "Extreme", ruedas = 2, material = "Carbono", 
                       cubierta = "SemiTacos", anchura_rueda = "MedioFina", amortiguacion_delantera = True)
    print(gravel)
    print("\n",mtb)
    print("\n",mono)
    print("\n", bCarretera)
    print("="*40)
    revisionPREuso(gravel)
    print("="*40)
    revisionPREuso(mtb)
    print("="*40)
    revisionPREuso(mono)
    print("="*40)
    revisionPREuso(bCarretera)
```
